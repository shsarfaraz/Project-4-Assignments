```python
import random
print("handout\nHigh Low Game\n")

round = 5

def main():
  print("Welcome to the High Low Game ðŸ˜Š")
  print("********************************")

  your_score = 0
  for i in range(round):
    print("Round", 1 + i)
    print("Your Score: " + str(your_score))

    computer_number = random.randint(1, 100)
    your_number = random.randint(1, 100)

    print("Your Number is: ", your_number)

    choice: str = input("Do you think your number is higher or lower than the computer's number? ")
    higher_and_correct : bool = choice == "higher" and your_number > computer_number
    lower_and_correct : bool = choice == "lower" and your_number < computer_number

    if higher_and_correct or lower_and_correct:
      print ("You'r right! The Computer\'s Number was ", computer_number)
      your_score += 1
    else:
      print("You're wrong! The Computer\'s Number was ", computer_number)

    print("Your score is now ",your_score)
    print()

  print("Thanks for Playing!")

if __name__ == "__main__":
  main()





```

```python
print("Intro to Python / handout.\n")
print("Planetary Weight Calculator\n")

print("Welcome to the Planetary Weight Calculator!")
print("******************************************")

def main():

  earth_weight = float(input("Enter your weight on Earth: "))
  gravity_ratios = {
    "Mercury": 0.38,
    "Venus": 0.91,
    "Mars": 0.38,
    "Jupiter": 2.34,
    "Saturn": 0.93,
    "Uranus": 0.92,
    "Neptune": 1.19
}

  print("\nSelect a planet to calculate your weight on: ")
  for planet in gravity_ratios:
    print(f" - {planet}")
  planet_choice = input("Enter the name of the planet: ").title()

  if planet_choice in gravity_ratios:
    new_weight = gravity_ratios[planet_choice]
    planetary_weight = earth_weight * new_weight
    print(f"Your weight on {planet_choice} is approximately {planetary_weight:.2f} kg.")
  else:
    print("Invalid Choice")

if __name__ == "__main__":
  main()
```

```python
print("Lists and Dicts/handout\n")
print("List Practice\n")

# Corrected list to separate items
my_list = ["Apple", "Banana", "Cherry", "Mango", "Orange", "Pear", "Peach"]
print("Current List:", my_list)

def access_element(my_list, index):
    """Returns the element at the given index in the list."""
    if 0 <= index < len(my_list):
        return f"Element at index {index}: {my_list[index]}"
    else:
        return "Index out of range"

def modify_element(my_list, index, new_value):
    """Modifies the element at the given index in the list."""
    if 0 <= index < len(my_list):
        old_value = my_list[index]
        my_list[index] = new_value
        return f"Element at index {index} modified from {old_value} to {new_value}"
    else:
        return "Index out of range"

def slice_list(my_list, start, end):
    """Returns a new list containing the elements from the start index to the end index."""
    if 0 <= start <= end < len(my_list):
        return f"Sliced List: {my_list[start:end]}"
    else:
        return "Invalid slice indices"

def list_game():
    print("\nWelcome to the List Manipulation Game!")

    while True:
        print("\nCurrent List:", my_list)

        print("Select an Operation: ")
        print("1. Access Element")
        print("2. Modify Element")
        print("3. Slice List")
        print("4. Quit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            index = int(input("Enter the index of the element you want to access: "))
            result = access_element(my_list, index)
            print(result)

        elif choice == "2":
            index = int(input("Enter the index of the element you want to modify: "))
            new_value = input("Enter the new value for the element: ")
            result = modify_element(my_list, index, new_value)
            print(result)

        elif choice == "3":
            start = int(input("Enter the start index for the slice: "))
            end = int(input("Enter the end index for the slice: "))
            result = slice_list(my_list, start, end)
            print(result)

        elif choice == "4":
            print("You exited the game. Thanks for playing!")
            break

        else:
            print("Invalid Choice, Please Enter a Number Between 1-4")

if __name__ == "__main__":
    list_game()
```
